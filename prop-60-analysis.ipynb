```python
prop = props[props.prop_name =="PROPOSITION 060- ADULT FILMS. CONDOMS. HEALTH REQUIREMENTS. INITIATIVE STATUTE."]
```

```python
merged = pd.merge(prop, contribs, on="calaccess_committee_id")
```

```python
merged.committee_position.value_counts()
```

```python
merged.info()
```

```python
merged.amount.sum()
```

```python
merged.committee_position.value_counts()
```

```python
support = merged[merged.committee_position == "SUPPORT"]
```

```python
oppose = merged[merged.committee_position == "OPPOSE"]
```

```python
support.amount.sum()
```

```python
oppose.amount.sum()
```

```python
support.sort_values("amount", ascending=False)
```

The example below includes every donor, from those who only has a lastname to those who has fisrtname too

```python
merged.groupby(["contributor_lastname", "committee_position"]).amount.sum().reset_index().sort_values("amount", ascending=False)
```

However, here it is only displayed those donors who are individuals. The reason is that organizations don't have a first name and, as a consequence of the indications given in this groupby, they don't appear.

```python
merged.groupby(["contributor_firstname", "contributor_lastname", "committee_position"]).amount.sum().reset_index().sort_values("amount", ascending=False)
```
